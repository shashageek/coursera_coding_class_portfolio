```python
import pandas as pd
```

```python
return_m = pd.read_csv("data/Debt.csv",
                     header = 0, index_col = 0, parse_dates = True, na_values = -99.99
                     )
return_m.index = pd.to_datetime(return_m.index, format="%Y%m").to_period('M')
return_m.columns = return_m.columns.str.strip()
return_m = return_m[['US_AGG','GLOBAL_AGG','EMSov']]
```

```python
def drawdown(return_series: pd.Series): 
    """
    Takes a times series of asset returns
    Computes and returns a Dataframe that contains:
    the wealth index
    the previous peaks
    percent drawdowns
    """
    wealth_index = 1000 * (1+return_series).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdowns = (wealth_index - previous_peaks)/previous_peaks
    return pd.DataFrame({
        "Wealth": wealth_index,
        "Peaks":  previous_peaks,
        "Drawdown": drawdowns
    })
```

```python
drawdown(return_m["SP500"])[["Wealth","Peaks"]].plot()
```
