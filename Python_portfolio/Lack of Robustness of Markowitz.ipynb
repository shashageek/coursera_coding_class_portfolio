Lack of Robustness of the Markowitz Procedure, and the GMV

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
import pandas as pd
import edhec_risk_kit as erk

ind = erk.get_ind_returns()
er = erk.annualize_rets(ind["1996":"2000"],12)
cov = ind["1996":"2000"].cov()
```

```python
l = ["Food","Steel"]
erk.msr(0.1, er[l],cov.loc[l,l])
```
